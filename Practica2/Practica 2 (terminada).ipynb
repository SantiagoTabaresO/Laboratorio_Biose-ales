```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def impseq(n0, n1, n2):
#   Genera x(n) = delta(n-n0); n1 <= n <= n2
#	 ----------------------------------------------
    n = np.arange(n1,n2+1) # Se crea el vector de muestras
    x = (n-n0) == 0
    
    plt.stem(n, x, use_line_collection=True)
    plt.title(f'Función Impulso Unitario u(n-{n0})')
    plt.xlabel('n')
    plt.ylabel('x(n)')
    plt.grid(True)
    plt.show()
    
    return [x,n]
```

```python
impseq(2,0,4)
```

# 	Implemente una función en Python que permita generar una función escalón unitario, definida en un intervalo n_1≤n_0≤n_2

```python
import numpy as np
import matplotlib.pyplot as plt

def escalon_unitario(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n - n0) >= 0

    plt.stem(n, x, use_line_collection=True)
    plt.title(f'Función Escalón Unitario u(n-{n0})')
    plt.xlabel('n')
    plt.ylabel('x(n)')
    plt.grid(True)
    plt.show()
    
    return x, n
```

```python
escalon_unitario(3, 0, 10)
```

# 	Implemente una función en Python que permita generar una función rampa, definida en un intervalo n_1≤n_0≤n_2 con una pendiente m. 

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion_rampa(m, n0, n1, n2):
    n = np.arange(n1, n2+1)
    rampa = m * (n - n0) * ((n - n0) >= 0)
    
    plt.stem(n, rampa, use_line_collection=True)
    plt.title(f'Función Rampa con pendiente m={m}')
    plt.xlabel('n')
    plt.ylabel('rampa(n)')
    plt.grid(True)
    plt.show()
    
    return rampa, n
```

```python
funcion_rampa(2, 3, -5, 10)
```

# 3.	Genere las siguientes secuencias usando las funciones básicas de Python que se han presentado. Grafique los resultados.

# 	x_1 (n)= 3δ(n+1)+5δ(n+3)+3δ(n+2)+3δ(n)+ δ(n). ¿Cómo debe ser el vector de muestras?

```python
n1 = -10 
n2 = 10 

delta1, n = impseq(-1, n1, n2)
delta2, _ = impseq(0, n1, n2)
delta3, _ = impseq(2, n1, n2)
delta4, _ = impseq(3, n1, n2)
    
x1 = 3 * delta1 + 3 * delta2 + 3 * delta3 + 5 * delta4
    
plt.stem(n, x1, use_line_collection=True)
plt.title(r'$x_1(n) = 3\delta(n+1) + 5\delta(n+3) + 3\delta(n+2) + 3\delta(n) + \delta(n)$')
plt.xlabel('n')
plt.ylabel(r'$x_1(n)$')
plt.grid(True)
plt.show()
    
x1,n
```

# 	El siguiente sistema arroja una secuencia definida entre  -6≤n≤6
# X_2(n) = -2u(n-4) + 3u(n-2)

```python
def stepseq(n0, n1, n2):
    n = np.arange(n1, n2 + 1)
    x = (n - n0) >= 0
    return x, n

n1 = -6
n2 = 6

u1, n = stepseq(4, n1, n2)  # u(n-4)
u2, _ = stepseq(2, n1, n2)  # u(n-2)

X2 = -2 * u1 + 3 * u2

plt.stem(n, X2, use_line_collection=True)
plt.title(r'$X_2(n) = -2u(n-4) + 3u(n-2)$')
plt.xlabel('n')
plt.ylabel('X_2(n)')
plt.grid(True)
plt.show()
```

# 	El siguiente sistema arroja una secuencia definida entre  0≤n≤100
# X_3(n) = e^(0.01n) * sen(2 * pi * n)

```python
n = np.arange(0, 101)

X3 = np.exp(0.01 * n) * np.sin(2 * np.pi * n)

plt.stem(n, X3, use_line_collection=True)
plt.title(r'$X_3(n) = e^{0.01n} \cdot \sin(2\pi n)$')
plt.xlabel('n')
plt.ylabel('X_3(n)')
plt.grid(True)
plt.show()
```

# 	x_4 (n)= 2r(n+3)-r(n-2)-5u(n-3),-10≤n≤10

```python
n1 = -10
n2 = 10 

rampa1, n = funcion_rampa(1, -3, n1, n2)  # r(n+3)
rampa2, _ = funcion_rampa(1, 2, n1, n2)   # r(n-2)
escalon, _ = escalon_unitario(3, n1, n2)  # u(n-3)
    
x4 = 2 * rampa1 - rampa2 - 5 * escalon

plt.stem(n, x4, use_line_collection=True)
plt.title(r'$x_4(n) = 2r(n+3) - r(n-2) - 5u(n-3)$')
plt.xlabel('n')
plt.ylabel(r'$x_4(n)$')
plt.grid(True)
plt.show()
    
x4, n
```

# Punto 4

```python
n = np.arange(0, 18)

X = np.piecewise(n, [n <= 5, (n >= 6) & (n <= 11), (n >= 12) & (n <= 17)], 
                 [lambda n: n, lambda n: n - 5, lambda n: n - 10])

X_prime = np.diff(X, prepend=0)  # La derivada discreta

# Graficar ambas secuencias
plt.figure(figsize=(12, 6))

# Subplot 1: Secuencia original
plt.subplot(2, 1, 1)
plt.stem(n, X, use_line_collection=True)
plt.title('Secuencia original $X[n]$')
plt.xlabel('n')
plt.ylabel('X[n]')
plt.grid(True)

# Subplot 2: Derivada discreta de la secuencia
plt.subplot(2, 1, 2)
plt.stem(n, X_prime, use_line_collection=True)
plt.title('Derivada discreta $X\'[n]$')
plt.xlabel('n')
plt.ylabel('X\'[n]')
plt.grid(True)

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```

# Descripcion:
La secuencia X[n] muestra un comportamiento de rampa creciente en tres segmentos diferentes. En cada segmento, la rampa comienza desde cero y crece linealmente con 𝑛. La derivada discreta X′[n] es constante en los intervalos donde X[n] tiene una pendiente constante. Específicamente, X′[n]=1 en estos intervalos. En las transiciones entre los segmentos (específicamente en n=6 y n=12), se observa un pico positivo. Esto ocurre porque hay un salto en el valor de X[n] al cambiar de un segmento al siguiente.

En el caso continuo, la derivada de una rampa sería una constante en el intervalo donde la rampa tiene una pendiente constante. En el caso discreto, la derivada se representa como la diferencia finita entre valores consecutivos, lo que introduce saltos o discontinuidades en las transiciones entre segmentos de la secuencia. La derivada discreta no es exactamente la misma que la derivada continua debido a la naturaleza discreta de la secuencia. Mientras que en el continuo la derivada sería suave, en la versión discreta, las transiciones abruptas entre segmentos se traducen en picos o discontinuidades en la secuencia derivada.

```python
rampa_creciente = np.arange(0, 6)  # {0, 1, 2, 3, 4, 5}
rampa_decreciente = np.arange(5, -1, -1)  # {5, 4, 3, 2, 1, 0}
rampa_creciente_2 = np.arange(1, 6)  # {1, 2, 3, 4, 5}
secuencia_constante_5 = np.full(4, 5)  # {5, 5, 5, 5}
secuencia_constante_10 = np.full(4, 10)  # {10, 10, 10, 10}

# Concatenar todas las partes para formar x(n)
x = np.concatenate([
    rampa_creciente,
    rampa_decreciente,
    rampa_creciente_2,
    secuencia_constante_5,
    secuencia_constante_10
])

# Definir el intervalo de n
n = np.arange(len(x))

# Graficar la secuencia x(n)
plt.figure(figsize=(10, 6))
plt.stem(n, x, use_line_collection=True)
plt.title('Secuencia $x(n)$')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()

# Encontrar x_5(n) = 2x(n-4) + x(n)
x_shifted_4 = np.concatenate((np.zeros(4), x[:-4]))  # Desplazar x(n) por 4
x_5 = 2 * x_shifted_4 + x

# Graficar x_5(n)
plt.figure(figsize=(10, 6))
plt.stem(n, x_5, use_line_collection=True)
plt.title(r'$x_5(n) = 2x(n-4) + x(n)$')
plt.xlabel('n')
plt.ylabel('x_5(n)')
plt.grid(True)
plt.show()

# Encontrar x_6(n) en el intervalo -20 ≤ n ≤ 20
n_full = np.arange(-20, 21)
x_full = np.zeros(len(n_full))

# Ajustar la inserción de x(n) en x_full
start_idx = len(n_full) // 2 - len(x) // 2
x_full[start_idx:start_idx+len(x)] = x

x_shifted_2 = np.concatenate((np.zeros(2), x_full[:-2]))  # Desplazar x_full(n) por 2

# Calcular x_6(n)
x_6 = 0.001 * np.exp(0.5 * n_full) * x_full + 10 * np.sin(0.05 * np.pi * n_full) * x_shifted_2

# Graficar x_6(n)
plt.figure(figsize=(10, 6))
plt.stem(n_full, x_6, use_line_collection=True)
plt.title(r'$x_6(n) = 0.001e^{0.5n}x(n) + 10\sin(0.05\pi n)x(n+2)$')
plt.xlabel('n')
plt.ylabel('x_6(n)')
plt.grid(True)
plt.show()
```
