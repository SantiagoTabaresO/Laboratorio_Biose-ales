# j)	Construya la función  y[n]=sin⁡(π*0.12n) en el intervalo 0 <= n <= 100.

```python
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)

plt.plot(n, y)
plt.title(r'$y[n] = \sin(\pi \cdot 0.12 \cdot n)$')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)
plt.show()
```

# Construya otra senoidal y2[n]=cos⁡(2π*0.03n) 

```python
n = np.arange(0, 101)
y2 = np.cos(2 * np.pi * 0.03 * n)

plt.plot(n, y2, label=r'$y_2[n] = \cos(2\pi \cdot 0.03 \cdot n)$', color='orange')
plt.title(r'Gráfica de $y_2[n] = \cos(2\pi \cdot 0.03 \cdot n)$')
plt.xlabel('n')
plt.ylabel('y_2[n]')
plt.grid(True)
plt.legend()
plt.show()
```

# 	Genere una tercera señal que sea la suma de estas dos señales, es decir 
# s[n]=y[n]+y2[n] y una cuarta, que sea el producto de estas, t[n]=y[n].y2[n].


```python
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)
y2 = np.cos(2 * np.pi * 0.03 * n)
s = y + y2
t = y * y2

plt.figure(figsize=(10, 8))

# Graficar y[n]
plt.subplot(4, 1, 1)
plt.plot(n, y, label=r'$y[n] = \sin(\pi \cdot 0.12 \cdot n)$')
plt.title(r'Gráfica de $y[n]$, $y_2[n]$, $s[n]$ y $t[n]$')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)
plt.legend()

# Graficar y2[n]
plt.subplot(4, 1, 2)
plt.plot(n, y2, color='orange', label=r'$y_2[n] = \cos(2\pi \cdot 0.03 \cdot n)$')
plt.xlabel('n')
plt.ylabel('y_2[n]')
plt.grid(True)
plt.legend()

# Graficar s[n] = y[n] + y2[n]
plt.subplot(4, 1, 3)
plt.plot(n, s, color='green', label=r'$s[n] = y[n] + y_2[n]$')
plt.xlabel('n')
plt.ylabel('s[n]')
plt.grid(True)
plt.legend()

# Graficar t[n] = y[n] * y2[n]
plt.subplot(4, 1, 4)
plt.plot(n, t, color='red', label=r'$t[n] = y[n] \cdot y_2[n]$')
plt.xlabel('n')
plt.ylabel('t[n]')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# m)	Graficar en la misma figura las señales y[n] y y2[n], utilice leyenda y colores diferentes para cada señal. Asigne título a cada eje.

```python
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)
y2 = np.cos(2 * np.pi * 0.03 * n)
s = y + y2
t = y * y2

# Graficar ambas señales en la misma figura
plt.figure(figsize=(10, 6))
plt.plot(n, s, label=r'$s[n] = y[n] + y_2[n]$', color='green')
plt.plot(n, t, label=r'$t[n] = y[n] \cdot y_2[n]$', color='red')

plt.title('Gráfico de $s[n]$ y $t[n]$')
plt.xlabel('n (índice de la muestra)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

# Reto: Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
def NotasEstudiantes(notas):
    valores = list(notas.values())
    
    nota_min = np.min(valores)
    nota_max = np.max(valores)
    nota_media = np.mean(valores)
    desviacion_tipica = np.std(valores)
    
    resumen = pd.Series({
        'Nota Mínima': nota_min,
        'Nota Máxima': nota_max,
        'Nota Media': nota_media,
        'Desviación Típica': desviacion_tipica
    })
    
    return resumen
```

```python
# Ejemplo de uso
notas_alumnos = {
    'Juliana': 4.7,
    'María': 4.2,
    'Pedro': 3.8,
    'Ricardo': 4.4,
    'Luisa': 4.3,
    'Aureliano': 2.9
}

Notas = NotasEstudiantes(notas_alumnos)
print(Notas)
```
